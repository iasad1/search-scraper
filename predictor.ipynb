```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
df = pd.read_csv('final-data.csv',index_col=None)
df
```

```python
df['prop_price'] = df['prop_price'].str.replace(',','')
```

```python
df
```

```python
df['prop_size(lower)'] = df['prop_size(lower)'].str.replace(',','')
```

```python
df
```

```python
df['prop_price']= df['prop_price'].astype(int)
```

```python
df['prop_size(lower)']= df['prop_size(lower)'].astype(int)
```

```python
df.dtypes
```

```python
df['post_code'] = df['prop_desc'].str.rsplit(',').str[-1] 

```

```python
df['out_code'] = df['post_code'].str.rsplit(' ').str[1]
df
```

```python
df_revised = df[['city','prop_size(lower)','out_code','prop_price']]
df_revised
```

# Exploring data

```python
df_revised.shape
```

```python
# Finding which columns have missing data
df_revised.isnull().any()
```

```python
# Counting the number of missing values
df_revised.isnull().sum(axis = 0)
```

```python
df_prep = df_revised.dropna()
```

```python
df_prep.describe().astype(int)
```

```python
plt.hist(df_prep['prop_price'])
plt.show()
```

# Dealing with irrational and missing data

```python
# Get names of indexes for which column size_lower has value 0
indexNames = df_prep[df_prep['prop_size(lower)'] <= 1000 ].index
 
# Delete these row indexes from dataFrame
df_prep.drop(indexNames , inplace=True)
```

```python
# Get names of indexes for which column size_lower has value 0
indexNames = df_prep[df_prep['prop_price'] <= 10000 ].index
 
# Delete these row indexes from dataFrame
df_prep.drop(indexNames , inplace=True)
```

```python
df_prep.shape
```

```python
df_prep.describe().astype(int)
```

# Split data into input (independant) and output (dependant) variable

```python
df_prep
```

```python
X = df_prep.drop(columns='prop_price')
X
```

```python
y= df_prep['prop_price']
y
```

# Encoding with dummy variables

```python
X = pd.get_dummies(data=X, columns=['city', 'out_code'])
X
```

# Splitting the data into training and test sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state = 0)
```

```python
X_train
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

# Running regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

# Compare the prediction vs actual

```python
y_test
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred.astype(int)})
df
```

# Model evaluation

```python
from sklearn.model_selection import cross_val_score
```

```python
print(cross_val_score(regressor,X_train,y_train, cv=5))
```

```python
np.mean(cross_val_score(regressor,X_train,y_train, cv=5))
```

```python
regressor.score(X_test,y_test)
```

# Plotting Test vs prediction

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predictor(X), color = 'green')
plt.title('Actual vs prediction')
plt.xlabel('actual')
plt.ylabel('pred')
plt.show()
```
