```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
df_csv = pd.read_csv("prop_details.csv")
```

```python
df_csv
```

```python
df_preclean = df_csv[['city','prop_type','prop_size','prop_desc','prop_price']].copy()
```

```python
df_preclean['prop_price'] = df_preclean['prop_price'].str.replace('- £','')
```

```python
df_preclean
```

```python
df_preclean['prop_price'] = df_preclean['prop_price'].str.replace(',','')
```

```python
df_preclean
```

```python
df_preclean['prop_size'] = df_preclean['prop_size'].str.replace('Sq Ft','')
```

```python
df_preclean
```

```python
df_preclean = df_preclean.rename(columns={"prop_size": "prop_size(Sq Ft)"})
```

```python
df_preclean
```

```python
# Splitting the size column into lower and upper limit
new_preclean = df_preclean["prop_size(Sq Ft)"].str.split("to", n = 1, expand = True)
```

```python
# making separate lower limit from new data frame 
df_preclean["prop_size(lower) SQ Ft"]= new_preclean[0] 
```

```python
# making separate lower limit from new data frame 
df_preclean["prop_size(upper) SQ Ft"]= new_preclean[1] 
```

```python
df_preclean
```

```python
df_preclean.drop(columns='prop_size(Sq Ft)',inplace=True)
```

```python
df_preclean
```

```python
df_preclean = df_preclean.rename(columns={"prop_price": "prop_price(£)"})
```

```python
df_preclean
```

```python
df_preclean['prop_size(lower) SQ Ft'] = df_preclean['prop_size(lower) SQ Ft'].str.replace(',','')
```

```python
df_preclean['prop_size(upper) SQ Ft'] = df_preclean['prop_size(upper) SQ Ft'].str.replace(',','')
```

```python
df_preclean
```

```python
mapping = {1:'Office', 2:'Industrial', 3:'Land', 4:'Retail', 5:'Leisure', 6:'Other'}
df_preclean_final =df_preclean.replace({'prop_type': mapping})
```

```python
df_preclean_final
```

```python
df_preclean_final.describe()
```

```python
df_preclean_final['prop_price(£)'].isnull()
```

```python
price_median = df_preclean_final['prop_price(£)'].median()
```

```python
df_preclean_final['prop_price(£)'].fillna(price_median,inplace=True)

```

```python
df_preclean_final
```

```python
size_lower_median = df_preclean_final['prop_size(lower) SQ Ft'].median()
```

```python
df_preclean_final['prop_size(lower) SQ Ft'].fillna(size_lower_median, inplace = True)
```

```python
 df_preclean_final
```

```python
size_upper_median = df_preclean_final['prop_size(upper) SQ Ft'].median()
```

```python
df_preclean_final['prop_size(upper) SQ Ft'].fillna(size_upper_median, inplace = True)
```

```python
df_preclean_final
```

```python
df_preclean_final.describe()
```

```python
df_preclean_final_zero_missing = df_preclean_final
```

```python
df_preclean_final_zero_missing.to_csv("no-missing.csv",index=False)
```
