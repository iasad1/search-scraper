```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
df = pd.read_csv('no-missing.csv')
```

```python
df
```

## Encoding variables with dummies

```python
encoded_columns = pd.get_dummies(df['city'])
df = df.join(encoded_columns).drop('city', axis=1)
```

```python
df
```

```python
encoded_columns = pd.get_dummies(df['prop_type'])
df = df.join(encoded_columns).drop('prop_type', axis=1)
```

```python
df
```

## Splitting the encoded data into input and output variables

```python
X = df.iloc[:,2:12]
```

```python
X
```

```python
y = df.iloc[:,1]
```

```python
y
```

## Splitting the data into training and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state = 0)
```

## Feature scaling

Feature scaling is squared to standardize or normalise the data. (Remember scaling)

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```
